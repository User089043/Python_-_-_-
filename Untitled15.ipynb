```python
# Задание 4
# -----------------------------------------------------------------------------------------------------------
# Рассчитайте сколько сотрудников работает в каждом департаменте
# -----------------------------------------------------------------------------------------------------------
# Выполним расчет с оглядкой на столбец 'left' (Уволился ли сотрудник), предполагая что признак 1 означает ДА
# -----------------------------------------------------------------------------------------------------------
df2 = df[df.left != 1].groupby('department')['department'].count()
df2.plot(kind = 'bar', title = 'Численность сотрудников в разрезе департаментов', xlabel = 'Департамент', ylabel = 'Количество сотрудников')
plt.xticks(rotation = 90)
for i, x in enumerate(df2.values):
    plt.text(i, x, int(x), horizontalalignment = 'center', verticalalignment = 'bottom', fontdict = {'size':10})
plt.show()
```

```python
# Задание 5
# -----------------------------------------------
# Показать распределение сотрудников по зарплатам
# -----------------------------------------------
df3 = df.groupby('salary')['salary'].count()
df3.plot(kind = 'bar', title = 'Распределение сотрудников по уровню зарплаты', xlabel = 'Относительный уровень зарплаты', ylabel = 'Количество сотрудников', figsize=(6, 7))
plt.xticks(rotation = 0)
for i, x in enumerate(df3.values):
    plt.text(i, x, int(x), horizontalalignment = 'center', verticalalignment = 'bottom', fontdict = {'size':10})
plt.show()
```

```python
# Задание 6
# ------------------------------------------------------------------------------------
# Показать распределение сотрудников по зарплатам в каждом департаменте по отдельности
# ------------------------------------------------------------------------------------
df4 = df.groupby([df['department'],df['salary']])[['salary']].count()
df4.plot(kind = 'bar', title = 'Распределение сотрудников по уровню зарплаты в каждом департаменте', xlabel = 'Относительный уровень зарплаты в разрезе департаментов', ylabel = 'Количество сотрудников', figsize = (15, 7), legend = None)
for i, x in enumerate(df4.values):
    plt.text(i, x, int(x), horizontalalignment = 'center', verticalalignment = 'bottom', fontdict = {'size':8})
plt.show()
```

```python
# Задание 7
# ------------------------------------------------------------------
# Проверить гипотезу, что сотрудники с высоким окладом
# проводят на работе больше времени, чем сотрудники с низким
# окладом
#
# Построим график отражающий зависимость этих столбцов друг от друга
# ------------------------------------------------------------------
sns.scatterplot(data=df, x='salary', y='average_montly_hours');
```

```python
# -------------------------------------------------------------------
# Исходя из графика я бы сказал что однозначный ответ дать невозможно
# Попробуем рассчитать коэффициент корреляции Кенделла, который
# сможет дать нам ответ. Для этого нам необходимо скорректировать
# набо данных для дальнейших вычислений и заменить категориальные
# данные поля salary на числовые. Предлагаю низкий уровень зарплаты
# обозначить цифрой 20000, средний цифрой 80000, а высокий 200000
# и рассчитать коэффициент корреляции Кенделла
# -------------------------------------------------------------------
df5 = df[['salary', 'average_montly_hours']]
df6 = df5.replace(['low', 'medium', 'high'], [20000, 80000, 200000])
df6[['salary', 'average_montly_hours']].corr(method='kendall')
```

```python
# ---------------------------------------------------------------------
# Полученные значения отражают отсутствие взаимосвязи (практически).
# А зачит гипотезу, что сотрудники с высоким окладом проводят на работе
# больше времени, чем сотрудники с низким окладом опровергнута
# ---------------------------------------------------------------------
```

```python
# Задание 8
# ------------------------------------------------------------------------------------------------
# Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):
#   - Доля сотрудников с повышением за последние 5 лет
#   - Средняя степень удовлетворенности
#   - Среднее количество проектов
#
# Для подсчета нам будет необходимо выполнить раздленение набора данных на два:
#   - уволенные сотрудники (df7)
#   - работающие сотрудники (df8)
# ------------------------------------------------------------------------------------------------
df7 = df.loc[df['left'] == 1]
df8 = df.loc[df['left'] == 0]
print('Доля сотрудников с повышением за последние 5 лет:')
print('  - среди уволившихся сотрудников составляет: ', len(df7.loc[df7['promotion_last_5years'] == 1]) / len(df7) * 100)
print('  - среди работающих сотрудников составляет: ', len(df8.loc[df8['promotion_last_5years'] == 1]) / len(df8) * 100)
print('Средняя степень удовлетворенности:')
print('  - среди уволившихся сотрудников составляет: ', df7['satisfaction_level'].mean())
print('  - среди работающих сотрудников составляет: ', df8['satisfaction_level'].mean())
print('Среднее количество проектов:')
print('  - среди уволившихся сотрудников: ', df7['number_project'].mean())
print('  - среди работающих сотрудников', df8['number_project'].mean())
```
